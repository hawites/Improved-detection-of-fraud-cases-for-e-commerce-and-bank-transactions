```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set display options
pd.set_option('display.max_columns', None)
sns.set(style='whitegrid')

# Load e-commerce fraud data
fraud_df = pd.read_csv('../data/Fraud_Data.csv')

# Load IP address to country mapping
ip_df = pd.read_csv('../data/IpAddress_to_Country.csv')

# Load credit card transaction data
credit_df = pd.read_csv('../data/creditcard.csv')

print("Fraud Data shape:", fraud_df.shape)
print("IP to Country shape:", ip_df.shape)
print("Credit Card Data shape:", credit_df.shape)

# Display first few rows
fraud_df.head()

# Check for null values
print("Missing values in Fraud Data:\n", fraud_df.isnull().sum())
print("Missing values in IP Data:\n", ip_df.isnull().sum())
print("Missing values in Credit Data:\n", credit_df.isnull().sum())

# Check for duplicates
print("Duplicates in Fraud Data:", fraud_df.duplicated().sum())
print("Duplicates in Credit Card Data:", credit_df.duplicated().sum())

# Data types
fraud_df.dtypes

fraud_df.describe(include='all').T
credit_df.describe().T

# For e-commerce data
print(fraud_df['class'].value_counts(normalize=True))

# For credit card data
print(credit_df['Class'].value_counts(normalize=True))

# Plot imbalance
sns.countplot(x='class', data=fraud_df)
plt.title("E-commerce Fraud Class Distribution")
plt.show()

sns.countplot(x='Class', data=credit_df)
plt.title("Credit Card Fraud Class Distribution")
plt.show()
```
